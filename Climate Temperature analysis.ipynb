```python
mask = 'Lithuania'
lithuania = countrytemp[countrytemp['Country'].str.contains(mask)]
```

```python
lithuania.head()
```

```python
lithuania.count()
```

```python
#checking for null values
lithuania.isnull().any()
```

```python
#Removing null values
lithuania = lithuania.dropna(how='any',axis=0)
```

```python
lithuania.head()
```

```python
#setting datetime to be index
lithuania = lithuania.set_index('dt')
```

```python
#Turnining into series, keeping only one Column Avg Temp
lithuania_average = lithuania['AverageTemperature']
lithuania_average.head()
```

```python
#Getting average temp per year
lithuania_year_averages = lithuania_average.resample('A').mean()
lithuania_year_averages.head()
```

```python
#Getting value counts for years
lithuania_value_counts = lithuania_average.resample('A').nunique()
lithuania_value_counts = lithuania_value_counts.rename('Counts')
lithuania_value_counts.head()
```

```python
#Adding new column to series
lithuania_concated = pd.concat([lithuania_year_averages, lithuania_value_counts], axis=1)
lithuania_concated.head()
```

```python
#Removing years with less than 12 months data
lithuania_data = lithuania_concated[lithuania_concated['Counts'] > 11]
lithuania_data.head()
```

```python
#Data for 260 years, clean and ready for plotting
lithuania_data.count()
```

```python
plt.figure(figsize=(15,15))
#plt.scatter(lithuania_data.index, lithuania_data.AverageTemperature)
plt.plot(lithuania_data.index, lithuania_data.AverageTemperature)
plt.title('Lithuania average tempartures for 260 Years',fontsize=20)
plt.ylabel('Temperature', fontsize=15)
```
